doc file to docx file

```python
import os
import win32com.client

def convert_doc_to_docx(doc_path, docx_path):
    word = win32com.client.Dispatch("Word.Application")
    try:
        doc = word.Documents.Open(doc_path)
        doc.SaveAs(docx_path, 16)  # 16 corresponds to the .docx format
        doc.Close()
    except Exception as e:
        print(f"Error converting {doc_path} to {docx_path}: {e}")
    finally:
        # Make sure to release the Word application properly
        if 'word' in locals():
            word.Quit()

def convert_docs_to_docx(src_folder, docx_folder):
    # Create the target folder if it doesn't exist
    if not os.path.exists(docx_folder):
        os.makedirs(docx_folder)

    # Find all .doc and .DOC files in the source folder
    doc_files = [f for f in os.listdir(src_folder) if f.lower().endswith(('.doc', '.docx'))]

    for doc_file in doc_files:
        doc_path = os.path.join(src_folder, doc_file)
        docx_path = os.path.join(docx_folder, os.path.splitext(doc_file)[0] + '.docx')

        # Convert .doc to .docx
        convert_doc_to_docx(doc_path, docx_path)
        print(f"Converted {doc_path} to {docx_path}")

if __name__ == "__main__":
    # Replace 'input_folder' with the path to your folder containing DOC files
    source_folder = r'D:\Work\IIT Bhilai\Internship\Specifications\series_25\docs'

    # Replace 'output_folder' with the desired path for the output DOCX files
    target_folder = r'D:\Work\IIT Bhilai\Internship\Specifications\series_25\docx'

    # Convert all .doc and .DOC files to .docx in the specified folder
    convert_docs_to_docx(source_folder, target_folder)
```

docx to pdf

```python
import os
import comtypes.client

def convert_docx_to_pdf(docx_path, pdf_path):
    try:
        word = comtypes.client.CreateObject("Word.Application")
        doc = word.Documents.Open(docx_path)
        doc.SaveAs(pdf_path, FileFormat=17)  # 17 corresponds to the PDF format
        doc.Close()
    except Exception as e:
        print(f"Error converting {docx_path} to {pdf_path}: {e}")
    finally:
        # Make sure to release the Word application properly
        if 'word' in locals():
            word.Quit()

def convert_docx_files_to_pdf(src_folder, pdf_folder):
    # Create the target folder if it doesn't exist
    if not os.path.exists(pdf_folder):
        os.makedirs(pdf_folder)

    # Find all DOCX files in the source folder
    docx_files = [f for f in os.listdir(src_folder) if f.lower().endswith('.docx')]

    for docx_file in docx_files:
        docx_path = os.path.join(src_folder, docx_file)
        pdf_path = os.path.join(pdf_folder, docx_file.replace('.docx', '.pdf'))

        # Convert .docx to .pdf
        convert_docx_to_pdf(docx_path, pdf_path)
        print(f"Converted {docx_path} to {pdf_path}")

if __name__ == "__main__":
    # Replace 'input_folder' with the path to your folder containing DOCX files
    source_folder = r'D:\Work\IIT Bhilai\Internship\Specifications\series_33\docx'

    # Replace 'output_folder' with the desired path for the output PDF files
    target_folder = r'D:\Work\IIT Bhilai\Internship\Specifications\series_33\pdf'

    # Convert all .docx files to .pdf in the specified folder
    convert_docx_files_to_pdf(source_folder, target_folder)
```
